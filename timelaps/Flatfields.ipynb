```python
import scipy
from scipy.misc import imread
import glob
```

```python
flats = [imread("flat-%i000.jpg"%i)
         for i in range(1,14)]
img = flats[0]
```

```python
figure()
imshow(img)
```

```python
def to_linear(img):
    compressed = img/255.0
    low = compressed <= (0.018*4.5)
    linear = (compressed/1.099)**(1.0/0.45)
    linear[low] = compressed[low]/4.5
    return linear    
def to_gamma(lin):
    low = lin<=0.018
    gamma = 1.099*(lin**0.45)
    gamma[low] = lin[low]*4.5
    return (gamma.clip(0, 1)*255).astype(numpy.uint8)
```

```python
figure()
imshow(to_linear(img), vmin=0, vmax=1)
```

```python
#Prepare azimuthal averaging
heigth, width = img.shape[:2]
y, x = numpy.ogrid[:heigth, :width]
y -= heigth//2
x -= width//2
radius = numpy.sqrt(x*x + y*y)
lin = to_linear(img)
hwr = numpy.histogram(radius.ravel(), bins=100, weights=(lin[:, :, 0]).ravel())[0]
hwg = numpy.histogram(radius.ravel(), bins=100, weights=(lin[:, :, 1]).ravel())[0]
hwb = numpy.histogram(radius.ravel(), bins=100, weights=(lin[:, :, 2]).ravel())[0]
hu = numpy.histogram(radius.ravel(), bins=100)
pos = 0.5*(hu[1][1:]+hu[1][:-1])
r = hwr/hu[0]
g = hwg/hu[0]
b = hwb/hu[0]
```

```python
figure()
plot(pos, r, "-r")
plot(pos, g, "-g")
plot(pos, b, "-b")
```

```python
fig, ax =subplots(len(flats), figsize=(10, 80))
for i,img in enumerate(flats):
    lin = to_linear(img)
    hwr = numpy.histogram(radius.ravel(), bins=100, weights=(lin[:, :, 0]).ravel())[0]
    hwg = numpy.histogram(radius.ravel(), bins=100, weights=(lin[:, :, 1]).ravel())[0]
    hwb = numpy.histogram(radius.ravel(), bins=100, weights=(lin[:, :, 2]).ravel())[0]
    ax[i].plot(pos, hwr/hu[0], "-r")
    ax[i].plot(pos, hwg/hu[0], "-g")
    ax[i].plot(pos, hwb/hu[0], "-b")
    ax[i].set_title("exposure= %i000 us"%((1+i)))
    

```

```python
valid = flats[:9]
fig, ax =subplots(len(valid)+1, figsize=(10, 40))
red = []
green = []
blue = []
for i,img in enumerate(valid):
    lin = to_linear(img)
    lin.shape = -1,3
    avg = lin.mean(axis=0)
    hwr = numpy.histogram(radius.ravel(), bins=100, weights=(lin[:, 0]/avg[0]))[0]
    hwg = numpy.histogram(radius.ravel(), bins=100, weights=(lin[:, 1]/avg[1]))[0]
    hwb = numpy.histogram(radius.ravel(), bins=100, weights=(lin[:, 2]/avg[2]))[0]
    red.append(hwr/hu[0])
    green.append(hwg/hu[0])
    blue.append(hwb/hu[0])
    ax[i].plot(pos, red[-1], "-r")
    ax[i].plot(pos, green[-1], "-g")
    ax[i].plot(pos, blue[-1], "-b")
    ax[i].set_title("exposure= %i000 us"%((1+i)))
ared = numpy.vstack(red).mean(axis=0)
agreen = numpy.vstack(green).mean(axis=0)
ablue = numpy.vstack(blue).mean(axis=0)
ax[-1].plot(pos, ared, "-r")
ax[-1].plot(pos, agreen, "-g")
ax[-1].plot(pos, ablue, "-b")
ax[-1].set_title("Average over valid curves")
```

```python
flatfield = numpy.empty((heigth, width, 3), dtype=numpy.float32)
```

```python
posex = numpy.concatenate(([0], pos, [radius.max()]))
aredex = numpy.concatenate(([ared[0]], ared, [ared[-1]]))
agreenex = numpy.concatenate(([agreen[0]], agreen, [agreen[-1]]))
ablueex = numpy.concatenate(([ablue[0]], ablue, [ablue[-1]]))
flatfield[:,:,0] = numpy.interp(radius, posex, aredex)
flatfield[:,:,1] = numpy.interp(radius, posex, agreenex)
flatfield[:,:,2] = numpy.interp(radius, posex, ablueex)
```

```python
def correct(img):
    return to_gamma(to_linear(img)/flatfield)
```

```python
fig, ax = subplots(9,2, figsize=(10,20))
for i,img in enumerate(flats[:9]):
    ax[i,0].imshow(flats[i])
    ax[i,0].set_title("raw %i000 us"%(i+1))
    ax[i,1].imshow(correct(flats[i]))
    ax[i,1].set_title("cor %i000 us"%(i+1))
```
