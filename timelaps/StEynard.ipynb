```python
ddir = "scaled1"
import os
if not os.path.exists(ddir):
    os.mkdir(ddir)
```

```python
cnt = 0
target=(1920,1080)
for fn, x, y, w, h in zip(images, x0, y0, w0, h0):
    im = Image.open(fn)
    dst = os.path.join(ddir,fn)
    box = (int(round(x)), int(round(y)), int(round(x+w)), int(round(y+h)))
    im.crop(box).resize(target, Image.ANTIALIAS).save(dst,quality=90)
    cnt+=1
    if cnt%10==0:
        print(fn)
print("Done")
```

```python
import scipy.misc
lum = []
thres = 0.2
for i,fn in enumerate(images[:-1]):
    img=scipy.misc.imread(fn,flatten=True)
    h,w = img.shape
    croped = (img[h//3:2*h//3,w//3:2*w//3]).ravel()
    croped.sort()
    l = len(croped)
    il=(croped[thres*l:(1-thres)*l]).mean()
    lum.append(il)
    if i%10==1: print(fn, il)
        
    
```

```python
figure()
plot(lum)
```

```python
figure()
plot(lum)
from scipy.ndimage import filters
smooth = filters.gaussian_filter1d(lum, 20)
plot(smooth)

#plot(lum/smooth)
```

```python
figure()
plot(smooth/lum)
```

```python
cnt = 0
target=(1920,1080)
for fn, x, y, w, h, c in zip(images, x0, y0, w0, h0, smooth/lum):
    img = scipy.misc.imread(fn)
    img2 = (img*c).clip(0,255).round().astype("int8")
    dst = os.path.join(ddir,fn)
    im = Image.fromarray(img2, "RGB")
    box = (int(round(x)), int(round(y)), int(round(x+w)), int(round(y+h)))
    im.crop(box).resize(target, Image.ANTIALIAS).save(dst,quality=90)
    cnt+=1
    if cnt%10==0:
        print(fn, c)
print("Done")
```
