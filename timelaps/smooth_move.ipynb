```python
data = json.load(open("position.json"))
```

```python
plot(data["dx"])
```

```python
pos_x = numpy.cumsum(data["dx"])
pos_y = numpy.cumsum(data["dy"])
pos = numpy.arange(pos_x.size)
print(pos_x.size)
from scipy.stats import linregress
slope_x,int_x,_,_,_ = linregress(pos, pos_x)
slope_y,int_y,_,_,_ = linregress(pos, pos_y)
theo_x = slope_x*pos+int_x
theo_y = slope_y*pos+int_y
plot(pos_x,label="meas")
plot(theo_x,label="exp")
legend()
delta_x = pos_x - theo_x
delta_y = pos_y - theo_y
```

```python
size = len(data["dx"])
width=1001
pos_x = numpy.cumsum(data["dx"]+[0]*width)
pos_y = numpy.cumsum(data["dy"]+[0]*width)
pos = numpy.arange(pos_x.size)
print(pos_x.size)
#from scipy.stats import linregress
#slope_x,int_x,_,_,_ = linregress(pos, pos_x)
#slope_y,int_y,_,_,_ = linregress(pos, pos_y)
from scipy.signal import gaussian, convolve
gau = gaussian(width,100)
gau/=gau.sum()
theo_x = convolve(pos_x,gau, mode="same")[:size]
theo_y = convolve(pos_y,gau, mode="same")[:size]
pos_x = pos_x[:size]
pos_y = pos_y[:size]
delta_x = pos_x - theo_x
delta_y = pos_y - theo_y
plot(pos_x,label="meas")
plot(theo_x,label="exp")
plot(delta_x, label="delta")
legend()
```

```python
print(delta_x.min(),delta_x.max())
print(delta_y.min(),delta_y.max())
```

```python
plot(pos_y,label="meas")
plot(theo_y,label="expect")
plot(delta_y,label="delta")
legend()
```

```python
ddir = "smooth"
import os
if not os.path.exists(ddir):
    os.mkdir(ddir)
```

```python
for fn,dx,dy in zip(images,delta_x,delta_y):
    im = Image.open(fn)
    dst = os.path.join(ddir,fn)
    box = (round(start[0]+dx), round(start[1]+dy), round(start[0]+dx)+target[0], round(start[1]+dy)+target[1])
    im.crop([int(i) for i in box]).save(dst)
    print(dst)
    
```
