```python
#This is 3 point measurement of the absolute rotation ...
angle_s = numpy.concatenate((numpy.linspace(-10,0,1988), numpy.linspace(0, 2.5, size-1988)))
print(len(angle_s)) 
plot(angle_s)
#angle_s*=numpy.pi/180.
```

```python
image_size = (2592, 1944)
from math import sin, cos, pi
def rotate(match_kp, angle1, angle2):
    ang1 = angle1 * pi / 180.0
    ang2 = angle2* pi / 180.0
    ca1 = cos(ang1)
    sa1 = sin(ang1)
    ca2 = cos(ang2)
    sa2 = sin(ang2)
    r1 = numpy.array([[ca1, sa1], [-sa1, ca1]])
    r2 = numpy.array([[ca2, sa2], [-sa2, ca2]])
    x1 = match_kp[:,0].x-image_size[0]/2.0
    y1 = match_kp[:,0].y-image_size[1]/2.0
    x2 = match_kp[:,1].x-image_size[0]/2.0
    y2 = match_kp[:,1].y-image_size[1]/2.0
    pos1 = numpy.dot(r1,numpy.vstack((x1,y1)))
    pos2 = numpy.dot(r2,numpy.vstack((x2,y2)))
    return numpy.median(pos2-pos1, axis=-1)
```

```python
angles = angle_s
#numpy.linspace(-6.44, 0, len(images))
idx = 1000
imshow(Image.open(images[idx]).rotate(angles[idx],resample=Image.BICUBIC))
```

```python
import h5py
import sift_pyocl as sift
match = sift.MatchPlan(devicetype='GPU')
h5 = h5py.File("keypoints.h5")
last = h5[images[0]].value
last_i = 0
rdx = []
rdy = []
for idx,fname in enumerate(images):
    kp = h5[fname].value
    m = match.match(last, kp)
    if len(m):
        x, y = rotate(m, angles[last_i], angles[idx])
    else:
        x = y = 0.0
    rdx.append(x)
    rdy.append(y)
    last_i = idx
    last = kp
    if idx%100==0: 
        print(fname)
rdx = numpy.array(rdx)
rdy = numpy.array(rdy)
print("Done")
```

```python
plot(rdx.cumsum())
plot(rdy.cumsum())
```

```python
size = len(rdx)#data["dx"])
sigma_v = 200
sigma_h = 100
width_h=8*sigma_h+1
width_v=8*sigma_v+1
pos_x = numpy.cumsum(list(rdx)+[0]*width_h)
pos_y = numpy.cumsum(list(rdy)+[0]*width_v)
from scipy.signal import gaussian, convolve
gau_v = gaussian(width_v,sigma_v)
gau_h = gaussian(width_h,sigma_h)
gau_h/=gau_h.sum()
gau_v/=gau_v.sum()
theo_x = convolve(pos_x, gau_h, mode="same")[:size]
theo_y = convolve(pos_y, gau_v, mode="same")[:size]
pos_x = pos_x[:size]
pos_y = pos_y[:size]
delta_x = pos_x - theo_x
delta_y = pos_y - theo_y
subplot(1,2,1)
plot(pos_x,label="meas")
plot(theo_x,label="exp")
plot(delta_x*10, label="delta*10")
title("X")
#legend()
subplot(1,2,2)
plot(pos_y,label="meas")
plot(theo_y,label="exp")
plot(delta_y*10, label="delta*10")
title("Y")
#legend()
print("Delta X: %s <--> %s"%(delta_x.min(), delta_x.max()))
print("Delta Y: %s <--> %s"%(delta_y.min(), delta_y.max()))
```

```python
%%cython
import numpy
def square_detection(data):
    cdef:
        int width, height, i, j
        int[:,::1] res, mask
    mask = numpy.ascontiguousarray(data, numpy.int32)
    res = numpy.zeros_like(mask)
    height = mask.shape[0]
    width = mask.shape[1]
    for i in range(height):
        res[i,0] = (mask[i,0] != 0)
    for j in range(width):
        res[0,j] = (mask[0,j] != 0)
    for i in range(1, height):
        for j in range(1, width):
            if mask[i,j] != 0:
                res[i,j] = 1 + min(res[i-1, j], res[i-1, j-1], res[i, j-1])
    return numpy.asarray(res)

#Some helper function for rectangles
cdef int getsize(int width, int height):
    "search for 16/9 images"
    return 16*width + 9*height

def rectangle_detection(data):
    cdef:
        int width, height, i, j, s, square, maxsquare, maxsize, rh, rw
        int[:,::1] sizes, squares, mask, widths, heights
        int[::1] height2width
    mask = numpy.ascontiguousarray(data, numpy.int32)
    squares = square_detection(mask)
    height = mask.shape[0]
    width = mask.shape[1]    
    sizes = numpy.zeros_like(mask)
    widths = numpy.zeros_like(mask)
    heights = numpy.zeros_like(mask)
    maxsquare = 0

    for i in range(height):
        for j in range(width):
            square = squares[i, j]
            sizes[i, j] = getsize(square, square)
            if square > maxsquare:
                maxsquare = square
    
    height2width = numpy.zeros(maxsquare+1, numpy.int32)    
    for i in range(height):
        height2width[:] = 0
        for j in range(width):
            square = squares[i, j]
            if square > 0:
                maxsize = sizes[i, j]
                for rh in range(square, 0, -1):
                    rw = height2width[rh]
                    rw = max(rw+1, square)
                    height2width[rh] = rw
                    size = getsize(rw, rh)
                    if size > maxsize:
                        maxsize = size
                        widths[i,j]=rw
                        heights[i,j]=rh
                    sizes[i,j] = maxsize
                height2width[square+1:] = 0
    return numpy.asarray(sizes), numpy.asarray(widths), numpy.asarray(heights)
```
